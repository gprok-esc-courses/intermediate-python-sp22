# Pandas example with UEFA data file

CSV file [download here](https://www.kaggle.com/datasets/sarangpurandare/uefa-champions-league-player-statistics)

```python
import pandas as pd
```

```python
df = pd.read_csv('uefa.csv')
```

```python
names = df['name']
```

```python
names.value_counts()
```

```python
df.loc[df['team']=='chelsea', 'name'].value_counts()
```

```python
df.loc[df['goals'] == df['goals'].max(), ['name', 'goals', 'season']]
```

```python
df.loc[(df['goals'] > 5) & (df['assists'] > 3), 'name']
```

```python
df_goals = df[['name', 'goals', 'team']]
```

```python
df_goals.info()
```

```python
df_goals.groupby(['name']).sum()
```

```python
matplotlib inline
```

```python
df_goals.groupby(['team']).sum().plot()
```
